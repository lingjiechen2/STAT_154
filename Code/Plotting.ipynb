```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
k_values = np.linspace(10, 100, 10)
recall = np.linspace(0.3, 0.9, len(k_values)) ** 2
top_k_em = recall - 0.2
top_1_em = top_k_em - 0.1

# Define a custom font properties
font_props = {'family': 'Times New Roman',
              'weight': 'normal',
              'size': 16,
              }

# Apply the custom font properties to all plots
plt.rc('font', **font_props)

# Plot the data
plt.figure(figsize=(8, 6))
plt.plot(k_values, recall, marker='x', label='Recall')
plt.plot(k_values, top_k_em, marker='+', label='Top k EM')
plt.plot(k_values, top_1_em, marker='o', label='Top 1 EM')

# Add titles and labels with the custom font properties
# plt.title('Model effectiveness with different numbers of retrieved paragraphs.', fontdict=font_props)
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.xlabel('k', fontdict=font_props)
plt.ylabel('Metric', fontdict=font_props)

# Customize x-axis and y-axis breaks (ticks)
plt.xticks(np.arange(0, 110, step=10))  # Set x-axis to show ticks every 10 steps
plt.yticks(np.arange(0, 1.1, step=0.1))  # Set y-axis to show ticks every 0.1 steps


# Add a legend with the custom font properties
plt.legend(prop=font_props)

# Customize the grid
plt.grid(True, linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```

```python
import json
import re
import jieba

def count_words(text):
    """ Counts words in a given text. """
    # words = re.findall(r'\b\w+\b', text)
    return len(text)

def average_word_count(data):
    """ Calculates the average word count for paragraphs and sentences. """
    total_paragraphs = len(data)
    total_sentences = 0
    total_words_in_paragraphs = 0
    total_words_in_sentences = 0

    for paragraph in data:
        total_words_in_paragraphs += count_words(paragraph)

        # Splitting paragraph into sentences
        sentences = re.split(r'[]', paragraph)
        sentences = [s.strip() for s in sentences if s.strip()]
        # print(sentences)
        total_sentences += len(sentences)

        for sentence in sentences:
            total_words_in_sentences += count_words(sentence)

    avg_words_per_paragraph = total_words_in_paragraphs / total_paragraphs
    avg_words_per_sentence = total_words_in_sentences / total_sentences

    return avg_words_per_paragraph, avg_words_per_sentence

with open('hw7_test.json', 'r') as f:
    data = json.load(f)
```
